```python
from math import log
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('bmh')

#Set up runtime comparisons
n = np.linspace(1, 10, 1000)
labels = ['Constant','Logarithmic','Linear','Log Linear','Quadratic','Cubic','Exponential']
big_o = [np.ones(n.shape),np.log(n),n,n*np.log(n),n**2,n**3,2**n]

# Plot setup
plt.figure(figsize=(12,10))
plt.ylim(0,100)

for i in range(len(big_o)):
    plt.plot(n,big_o[i],label = labels[i])


plt.legend(loc=0)
plt.ylabel('Relative Runtime')
plt.xlabel('n')
```

# Constant

```python
def BigO_Const(lit):
    print(lit[0]) #O(1)
```

```python
BigO_Const([0,1,2,3,4,5])
```

# Linear

```python
def BigO_Linear(lit):
    for val in range(lit): # O(n)
        print(val)
```

```python
BigO_Linear(10)
```

# Quadratic

```python
def BigO_Quadratic(lit):
    for val in range(lit):  #O(n^2)
        for val2 in range(val):
            print(val)
            print(val2)
```

```python
BigO_Quadratic(3)
```

# CUBIC

```python
def BigO_Cubic(lit):
    for val in range(lit):
        for val2 in range(val):
            for val3 in range(val2):
                print(val, val2, val3)
```

```python
BigO_Cubic(5)
```
